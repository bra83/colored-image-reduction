```python
from PIL import Image

def converter_para_cinza(imagem):

    largura, altura = imagem.size
    cinza = Image.new('L', (largura, altura))

    for x in range(largura):
        for y in range(altura):
            r, g, b = imagem.getpixel((x, y))
            cinza_pixel = int(0.299 * r + 0.587 * g + 0.114 * b)
            cinza.putpixel((x, y), cinza_pixel)

    return cinza

def converter_para_preto_e_branco(imagem_cinza, limiar=128):


    largura, altura = imagem_cinza.size
    preto_e_branco = Image.new('1', (largura, altura))

    for x in range(largura):
        for y in range(altura):
            cinza_pixel = imagem_cinza.getpixel((x, y))
            if cinza_pixel >= limiar:
                preto_e_branco.putpixel((x, y), 255)  # Branco
            else:
                preto_e_branco.putpixel((x, y), 0)  # Preto

    return preto_e_branco

# Carregando a imagem colorida
imagem = Image.open("/content/carro.jpeg")

# Convertendo para a escala de cinza
cinza = converter_para_cinza(imagem)

# Convertendo para preto e branco
preto_e_branco = converter_para_preto_e_branco(cinza)
```
